```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ddos_results = pd.read_csv('./all-raw-optimal-results.csv')
partB_results = pd.read_csv('/Users/jujohnson/git/CMS-Classification/post-processing/partB/all-raw-optimal-results.csv')
partD_results = pd.read_csv('/Users/jujohnson/git/CMS-Classification/post-processing/partD/all-raw-optimal-results.csv')
```

### Find All RUS and Baseline Results with Depth 2

```python
ddos_d2_results = ddos_results.loc[
    (ddos_results['layers'] == 2) &
    (ddos_results['method'] == 'rus')
]
ddos_d2_results.minority_size.sort_values().unique()
```

```python
partB_d2_results = partB_results.loc[
    (partB_results['layers'] == 2) &
    ((partB_results['method'].isin(['rus'])))
]
partB_d2_results.minority_size.sort_values().unique()
```

```python
partD_d2_results = partD_results.loc[
    (partD_results['layers'] == 2) &
    ((partD_results['method'].isin(['rus'])))
]
partD_d2_results.minority_size.sort_values().unique()
```

```python
print('Part D method count', len(partD_d2_results.method_name.unique()))
print('Part B method count', len(partB_d2_results.method_name.unique()))
print('DDOS method count', len(ddos_d2_results.method_name.unique()))
```

```python
print('Part D minority size count', len(partD_d2_results.minority_size.unique()))
print('Part B minority size count', len(partB_d2_results.minority_size.unique()))
print('DDOS minority size count', len(ddos_d2_results.minority_size.unique()))
```

### Plotting RUS AUC CIs

```python
all_sizes = np.sort(np.unique([
    *partB_d2_results.minority_size.unique(), 
    *partD_d2_results.minority_size.unique(),
    *ddos_d2_results.minority_size.unique()
]))
```

```python
all_sizes
```

```python
x = np.arange(0, len(all_sizes))

x_lookup = {}

for idx, size in enumerate(all_sizes):
    x_lookup[size] = x[idx]
```

```python
partB_d2_results['xtick'] = [x_lookup[val] for val in partB_d2_results['minority_size']]
partD_d2_results['xtick'] = [x_lookup[val] for val in partD_d2_results['minority_size']]
ddos_d2_results['xtick'] = [x_lookup[val] for val in ddos_d2_results['minority_size']]
```

```python
partB_d2_results['roc_auc_zero_mean'] = partB_d2_results['roc_auc'] - np.mean(partB_d2_results['roc_auc'])
partD_d2_results['roc_auc_zero_mean'] = partD_d2_results['roc_auc'] - np.mean(partD_d2_results['roc_auc'])
ddos_d2_results['roc_auc_zero_mean'] = ddos_d2_results['roc_auc'] - np.mean(ddos_d2_results['roc_auc'])
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(nrows=3,  ncols=1, sharex=True, figsize=(15, 10))
sns.lineplot(x='xtick', y='roc_auc', data=partB_d2_results, ax=ax1)
ax1.set_xticks(partB_d2_results['xtick'])
sns.lineplot(x='xtick', y='roc_auc', data=partD_d2_results, ax=ax2)
sns.lineplot(x='xtick', y='roc_auc', data=ddos_d2_results, ax=ax3)
plt.subplots_adjust(wspace=0, hspace=0.1)
```
